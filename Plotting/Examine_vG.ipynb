```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
pred_path = r'C:\Users\jv97\Desktop\github\Neutrino-Machine-Learning\predictions'
df = pd.read_csv(os.path.join(pred_path,'OscNext_Angle_m27_vG2_2jnn_test_predictions.csv'))
```

```python
df.head()
```

```python
df['r'] = np.sqrt(df.x_pred**2 + df.y_pred**2 + df.z_pred**2)
df['sig'] = np.sqrt(df.x_sig**2 + df.y_sig**2 + df.z_sig**2)
df['azimuth_pred'] = np.arctan2(df.y_pred,df.x_pred)
df['zenith_pred'] = np.arccos(df.z_pred/df.r)

def Omega(azp,zep,azt,zet):
    s1 = np.sin( zet + azt - azp )
    s2 = np.sin( zet - azt + azp )
    c1 = np.cos( zet - zep )
    c2 = np.cos( zet + zep )
    return np.arccos(0.5*np.abs(np.sin(zep))*( s1 + s2 ) + 0.5*(c1 + c2))
df['solid'] = Omega(df.azimuth_pred,df.zenith_pred,df.azimuth,df.zenith)
```

```python
fig, ax = plt.subplots()
ax.hist2d(df.solid,df.sig,bins=100,range=((0,np.pi),(0,np.quantile(df.sig,0.99))));

fig, ax = plt.subplots()
ax.hist2d(df.r,df.sig,bins=100,range=((0,1.5),(0,np.quantile(df.sig,0.99))));

fig, ax = plt.subplots()
ax.hist2d(df.solid,df.r,bins=100);
```
