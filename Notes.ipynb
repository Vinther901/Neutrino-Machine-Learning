# 1. Probabilistic angle regression for $\nu_\mu$

If one tries to predict azimuth and zenith for $\nu_\mu$ using the spherical likelihood loss function, the learning process keeps fluctuating and never converges. This is contrary to when the same loss function is employed for angle regression of $\mu$, there the use is succesful.

The problem seems to be that (atleast for now) the azimuth angle is near impossible to reconstruct for neutrinos. Thereby, using a loss function, where you transform the 2 angles into a single variable, namely the solid angle (angle between truth and prediction), means that the network cannot simply focus on getting the zenith angle correct while ignoring the azimuth.
