```python
df_truth = pd.read_csv(os.path.join(pred_path,'OscNext_Angle_m27_23ge_test_predictions.csv'))[['event_no','azimuth','zenith','energy_log10','pid','elasticity']]
```

```python
def preprocess(df):
    cols = df.columns
    if 'k_solid' in cols:
        kappa = df.k_solid**2 + 1e-5
        df['solid_sigma'] = np.sqrt(-2*np.log((1 + np.exp(-2*kappa))/(1 - np.exp(-2*kappa)) - 1/kappa))
    if 'k_az' in cols:
        from scipy.special import iv
        kappa =  df.k_az**2 + 1e-5
        df['azimuth_sigma'] = np.sqrt(-2*np.log(iv(1,kappa)/iv(0,kappa)))
    if 'k_ze' in cols:
        from scipy.special import iv
        kappa =  df.k_ze**2 + 1e-5
        df['zenith_sigma'] = np.sqrt(-2*np.log(iv(1,kappa)/iv(0,kappa)))
    if 'pred_sigma' in  cols:
        df.rename(columns={'pred_sigma':'solid_sigma'},inplace=True)
    if 'solid' not in cols:
        def Omega(azp,zep,azt,zet):
            s1 = np.sin( zet + azt - azp )
            s2 = np.sin( zet - azt + azp )
            c1 = np.cos( zet - zep )
            c2 = np.cos( zet + zep )
            return np.arccos(0.5*np.abs(np.sin(zep))*( s1 + s2 ) + 0.5*(c1 + c2))
        df['solid'] = Omega(df.azimuth_pred,df.zenith_pred,df_truth.azimuth,df_truth.zenith)
    cols = df.columns
    return df.iloc[:,cols.isin(['solid','azimuth_pred','zenith_pred','solid_sigma','azimuth_sigma','zenith_sigma'])]
```

```python
df_list = []
for csv in csv_list:
    df_list.append(preprocess(pd.read_csv(os.path.join(pred_path,csv))))
```

```python
labels = ['truth','SvM','bothvMF','2xPvM','MSE+MAE']
colors = ['k','r','g','b','orange']

fig, ax = plt.subplots(figsize=(14,7),ncols=2)

for x, label, color in zip([df_truth.azimuth]+[df.azimuth_pred for df in df_list],labels,colors):
    ax[0].hist(x%(2*np.pi),bins=100,range=(0,2*np.pi),label=label,color=color,histtype='step');
    
for x, label, color in zip([df_truth.zenith]+[df.zenith_pred for df in df_list],labels,colors):
    ax[1].hist(np.abs(x),bins=100,range=(0,np.pi),label=label,color=color,histtype='step');

ax[0].legend()
ax[1].legend()
```

```python
def plot_dist(x_list,bins=100,range=None, ax = None):
    if ax == None:
        fig, ax = plt.subplots()
    for x in x_list:
        ax.hist(x,bins=bins,range=None,histtype='step')

def plot_az(df,bins=100,range=None,ax=None):
    if ax == None:
        fig, ax = plt.subplots()
    plot_dist([df_truth.azimuth,(df.azimuth_pred)%(2*np.pi)],bins=100,range=None,ax=ax)
    
def plot_ze(df,bins=100,range=None,ax=None):
    plot_dist([df_truth.zenith,df.zenith_pred.abs()],bins=100,range=None,ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(14,7),ncols=2)

for df in df_list:
    plot_az(df,bins=100,range=(0,2*np.pi),ax=ax[0])
    plot_ze(df,bins=100,range=(0,np.pi),ax=ax[1])
```
